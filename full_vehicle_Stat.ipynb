```python
import pandas as pd
import os
file_path = r'PassengerVehicle_Stats'
file_list = [os.path.join(file_path, file) for file in os.listdir(file_path) if file.endswith('.csv')]
```

```python
dataframes = [pd.read_csv(file) for file in file_list]
vehicles_df = pd.concat(dataframes, ignore_index=True)
vehicles_df
```

```python
missing_values = vehicles_df.isnull().sum()
print(missing_values)
```

```python

Q1 = vehicles_df['Vehicle Model Year'].quantile(0.25)
Q3 = vehicles_df['Vehicle Model Year'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


vehicles_df = vehicles_df[(vehicles_df['Vehicle Model Year'] >= lower_bound) & (vehicles_df['Vehicle Model Year'] <= upper_bound)]
                          

```

```python
missing_values = vehicles_df.isnull().sum()
print(missing_values)
```

```python
vehicles_df.info()
```

```python
vehicles_df = vehicles_df.copy()

# Calculate the most frequent make
most_frequent_make = vehicles_df['Vehicle Make'].mode()[0]

# Fill missing values with the most frequent make
vehicles_df['Vehicle Make'] = vehicles_df['Vehicle Make'].fillna(most_frequent_make)
```

only 9 missing values have this 'Vehicle Make' coloum. It is a a small percentage compared to the rest of the dataset. So we replace missing values with the most frequent value (mode) for consistency.

```python
vehicles_df.loc[:, 'Vehicle Color'] = vehicles_df['Vehicle Color'].fillna('Unknown')
vehicles_df
```

#61 missing values are in the vehicle colour column. In this case vehicle's type or usage doesn’t depend on colour. so we can fill this missing cell using 'unknown' keyword. 

```python
vehicles_df = vehicles_df.copy()
vehicles_df.loc[:, 'City'] = vehicles_df['City'].fillna(most_frequent_make)
```

```python
vehicles_df = vehicles_df.dropna() 
```

#have 54 missing data in the 'Vehicle Model' cell. it is a small portion compare with the whole dsta in the column.
```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming agg_results is a DataFrame from your groupby operation
agg_results = vehicles_df.groupby(['Status', 'Vehicle Fuel Source']).agg(
    Count=('Record ID', 'count'),
    Avg_Model_Year=('Vehicle Model Year', 'max')
).reset_index()

# Plot configuration
fig, ax1 = plt.subplots(figsize=(12, 6))

# Barplot for Count
sns.barplot(
    data=agg_results,
    x='Vehicle Fuel Source',
    y='Count',
    hue='Status',
    ax=ax1
)
ax1.set_title('Vehicle Counts by Fuel Source and Status', fontsize=16)
ax1.set_ylabel('Count', fontsize=12)
ax1.set_xlabel('Vehicle Fuel Source', fontsize=12)
ax1.legend(title='Status')

# Secondary axis for Avg_Model_Year
ax2 = ax1.twinx()
sns.lineplot(
    data=agg_results,
    x='Vehicle Fuel Source',
    y='Avg_Model_Year',
    hue='Status',
    marker='o',
    ax=ax2,
    legend=False
)
ax2.set_ylabel('Average Model Year', fontsize=12)
ax2.grid(False)

# Show plot
plt.tight_layout()
plt.show()
```

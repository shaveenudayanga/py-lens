```python
import pandas as pd
import os
file_path = r'PassengerVehicle_Stats'
file_list = [os.path.join(file_path, file) for file in os.listdir(file_path) if file.endswith('.csv')]
```

```python
missing_values = vehicles_df.isnull().sum()
print(missing_values)
```

```python

Q1 = vehicles_df['Vehicle Model Year'].quantile(0.25)
Q3 = vehicles_df['Vehicle Model Year'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


vehicles_df = vehicles_df[(vehicles_df['Vehicle Model Year'] >= lower_bound) & (vehicles_df['Vehicle Model Year'] <= upper_bound)]
                          

```

```python
missing_values = vehicles_df.isnull().sum()
print(missing_values)
```

```python
vehicles_df.info()
```

```python
vehicles_df = vehicles_df.copy()

# Calculate the most frequent make
most_frequent_make = vehicles_df['Vehicle Make'].mode()[0]

# Fill missing values with the most frequent make
vehicles_df['Vehicle Make'] = vehicles_df['Vehicle Make'].fillna(most_frequent_make)
```

only 9 missing values have this 'Vehicle Make' coloum. It is a a small percentage compared to the rest of the dataset. So we replace missing values with the most frequent value (mode) for consistency.

```python
vehicles_df.loc[:, 'Vehicle Color'] = vehicles_df['Vehicle Color'].fillna('Unknown')
vehicles_df
```

61 missing values are in the vehicle colour column. In this case vehicle's type or usage doesn’t depend on colour. so we can fill this missing cell using 'unknown' keyword. 

```python
vehicles_df = vehicles_df.copy()
vehicles_df.loc[:, 'City'] = vehicles_df['City'].fillna(most_frequent_make)
```

```python
vehicles_df = vehicles_df.dropna() 
```

There are 54 missing data entries in the 'Vehicle Model' column. This is a small portion compared to the entire dataset in the column. We could assign some random values to these empty cells, but since this data is important, doing so might compromise the accuracy of our analysis. Therefore, we believe the most accurate approach is to delete all rows with empty cells in this column. 

```python
missing_values = vehicles_df.isnull().sum()
print(missing_values)
```
